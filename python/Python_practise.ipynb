Character frequency Counter


```python
sentence = input("Enter the string:")
freq ={}

for ch in sentence:
  if ch!= " ":
    freq[ch]=freq.get(ch,0)+1
print("Frequency is: ",freq)
```

Even- Odd Using Function

```python
def is_even(num):
  return  num %2 ==0

num = int(input("Enter the number:"))
if is_even(num):
  print("Even")
else:
  print("Odd")
```

List + Function

```python
def sum_list(nums):
    sum =0
    for num in nums:
      sum = sum +num
    return sum
nums = [1,2,3,4,5,6,7,8]

print("The sum is:",sum_list(nums))
```

List Comprehension

```python
nums= [2,4,6,7,8,10]
even_nums = []
for num in nums:
  if num%2==0:
    even_nums.append(num)
print(even_nums)
```

```python
def char_count(s):
  freq={}
  for ch in s:
    if ch!=" ":
     freq[ch]=freq.get(ch,0)+1
  return freq

sentence = input("Enter a string: ")
print(char_count(sentence))
```

```python
import string
def word_count(s):
  freq={}
  s_clean = s.lower().translate(str.maketrans('', '', string.punctuation))
  words = s_clean.split()
  for word in words:
     freq[word]=freq.get(word,0)+1
  return freq

sentence = input("Enter a string: ")
print(word_count(sentence))
```

```python
def second_largest(nums):
    first = second = float('-inf')
    for num in nums:
        if num > first:
            second = first
            first = num
        elif first > num > second:
            second = num
    return second

nums = [12, 20, 20, 8, 5]
print(second_largest(nums))
```
