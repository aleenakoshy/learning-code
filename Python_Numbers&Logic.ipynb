Even-Odd

```python
def check_even_odd(n):
 return  n%2==0

n = int(input("Enter the number:"))
if check_even_odd(n):
   print("The number is even")
else:
   print("The number is odd")
```

Prime-Number CHECK

```python
n =int(input("enter the number:"))
if n<2:
  print("Not prime")
else:
  is_prime = True
  for i in range(2, n):
    if n%i ==0:
      is_prime= False
      break
  if is_prime:
    print("Prime")
  else:
    print("Not prime")
```

Factorial Of  A number

```python
n = int(input("Enter the number:"))
factorial =1
for i in range(1,n+1):
  factorial= factorial *i
print("The factorial of the number is", factorial)
```

Largest In a list


```python
nums =[12,20,8,5]
largest = nums[0]
for num in nums:
  if num>largest:
    largest = num
print("The largest number is", largest)
```

Smallest in a List

```python
nums =[12,20,8,5]
smallest = nums[0]
for num in nums:
  if num<smallest:
    smallest = num
print("The smallest number is", smallest)
```

Second Largest

```python
nums = [12, 20, 20, 8, 5]
largest = second = float('-inf')

for num in nums:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num

print("Second largest number is:", second)
```

Reverse a number

```python
n = int(input("Enter the number:"))
rev = 0

while n>0:
  digit = n%10
  rev = rev*10 +digit
  n = n//10
print("The reverse is", rev)
```
